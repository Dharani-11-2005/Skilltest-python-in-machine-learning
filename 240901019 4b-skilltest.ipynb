```python
import pandas as pd

# Creating dictionary data
data = {
'Product': ['Laptop','Smartphone','Tablet','Monitor','Keyboard'],
'Price': [1200,300,150,250,50],
'Quantity': [10,20,15,25,50]
}

# Converting dictionary to DataFrame
df = pd.DataFrame(data)

print(df)

# Operation 1: Calculate average price
print("Average Price:", df['Price'].mean())

# Operation 2: Find maximum price
print("Maximum Price:", df['Price'].max())

# Operation 3: Find minimum price
print("Minimum Price:", df['Price'].min())

# Operation 4: Calculate total quantity of products
print("Total Quantity:", df['Quantity'].sum())

# Operation 5: Create new column for total stock value
df['Total_Value'] = df['Price'] * df['Quantity']

print(df)
```
