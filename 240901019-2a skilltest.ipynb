```python
import numpy as np

steps = np.array([4000, 5500, 7000, 8500, 9200, 6100, 7800])

total = np.sum(steps)
average = round(np.mean(steps), 2)
maximum = np.max(steps)
minimum = np.min(steps)

high_activity = steps[steps > 8000]

print("Total Steps:", total)
print("Average Steps:", average)
print("Maximum Steps:", maximum)
print("Minimum Steps:", minimum)
print("High Activity Days:", high_activity)
```
