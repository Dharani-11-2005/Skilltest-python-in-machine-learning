```python
import pandas as pd

data = {
'Participant': ['P1','P1','P2','P2','P3','P3'],
'Electrode': ['E1','E2','E1','E2','E1','E2'],
'Signal': [12,15,10,18,14,16]
}

df = pd.DataFrame(data)

print("EEG Data")
print(df)

print("Mean:", round(df['Signal'].mean(),3))
print("Variance:", round(df['Signal'].var(),3))
print("Peak Value:", df['Signal'].max())

print("Mean Signal by Participant")
print(df.groupby('Participant')['Signal'].mean())
```
