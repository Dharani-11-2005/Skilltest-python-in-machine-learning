```python
import numpy as np

stock1 = np.array([[100,105,98],
                   [102,108,101],
                   [104,110,103],
                   [106,112,105],
                   [108,115,107]])

stock2 = np.array([[95,100,90],
                   [97,102,92],
                   [99,104,95],
                   [101,106,97],
                   [103,108,99]])

sum_matrix = stock1 + stock2
diff_matrix = stock1 - stock2

print("Element-wise Sum:")
print(sum_matrix)

print("Element-wise Difference:")
print(diff_matrix)
```
