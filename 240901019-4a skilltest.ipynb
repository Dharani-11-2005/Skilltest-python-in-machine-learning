```python
import numpy as np

rain = np.array([12, 15, 10, 18, 20, 14, 16, 22])

avg = round(np.mean(rain),2)
maximum = np.max(rain)
minimum = np.min(rain)

heavy_rain = rain[rain > 18]

print("Average Rainfall:", avg)
print("Maximum Rainfall:", maximum)
print("Minimum Rainfall:", minimum)
print("Heavy Rainfall Days:", heavy_rain)
```
