```python
import numpy as np

scores = np.array([78,45,67,89,56,92,34,71,63,84,55,48,90,73,60,52,41,66,79,88])

average = np.mean(scores)
highest = np.max(scores)
lowest = np.min(scores)

print("Average Score:", average)
print("Highest Score:", highest)
print("Lowest Score:", lowest)

pass_students = scores[scores >= 50]
fail_students = scores[scores < 50]

print("Passed:", len(pass_students))
print("Failed:", len(fail_students))

grades = []

for s in scores:
    if s >= 90:
        grades.append("A")
    elif s >= 80:
        grades.append("B")
    elif s >= 70:
        grades.append("C")
    elif s >= 60:
        grades.append("D")
    else:
        grades.append("F")

print("Grades:", grades)
```
